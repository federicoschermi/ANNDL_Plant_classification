```python
import keras_cv   
import keras  
import tensorflow as tf  
import tensorflow_datasets as tfds  
from matplotlib import pyplot as plt 
from tensorflow.keras import regularizers
import random
import warnings
import os
from tensorflow.keras.callbacks import EarlyStopping

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
warnings.filterwarnings('ignore')

#using multi GPU
strategy = tf.distribute.MirroredStrategy()  

# Dataset hyperparameters
IMAGE_SIZE = 96
IMAGE_CHANNELS = 3
NUM_CLASSES = 2

# Algorithm hyperparameter
UNLABELED_BATCH_SIZE = 1024
LABELED_BATCH_SIZE = 128
TEST_BATCH_SIZE = 128
PROJECTION_WIDTH = 128
TEMPERATURE = 0.25


# Stronger augmentations for contrastive
CONTRASTIVE_AUGMENTATION = {
    "height_factor":0.6,
    "width_factor":0.6,
    "t_h":1,
    "t_w":1,
    "color_jitter_rate": 0.8,
    "brightness_factor": 0.2,
    "contrast_factor": 0.8,
    "saturation_factor": (0.3, 0.7),
    "hue_factor": 0.2,
    "rotation_rate":0,
    "zoom":(0,0),
    "gb1":6,
    "gb2":1,
}

# Weaker ones for supervised training
CLASSIFICATION_AUGMENTATION = {
    "height_factor":0.4,
    "width_factor":0.4,
    "t_h":1,
    "t_w":1,
    "color_jitter_rate": 0.05,
    "brightness_factor": 0.1,
    "contrast_factor": 0.1,
    "saturation_factor": (0.1, 0.1),
    "hue_factor": 0.2,
    "rotation_rate": 0,
    "zoom":(0,0),
    "gb1":3,
    "gb2":0.5,

}

AUTOTUNE = tf.data.AUTOTUNE

SEED = 42
DIR = "/kaggle/input/training-data-final/training_data_final"
def prepare_dataset():
    unlabeled_train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
        DIR,
        labels=None,
        class_names=None,
        color_mode="rgb",
        batch_size=UNLABELED_BATCH_SIZE,
        image_size=(96, 96),
        shuffle=True,
        seed=SEED,
        validation_split=0.2,
        subset="training",
        interpolation="bilinear",
        follow_links=False,
    )

    labeled_train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
        DIR,
        labels="inferred",
        label_mode="binary",
        class_names=None,
        color_mode="rgb",
        batch_size=LABELED_BATCH_SIZE,
        image_size=(96, 96),
        shuffle=True,
        seed=SEED,
        validation_split=0.2,
        subset="training",
        interpolation="bilinear",
        follow_links=False,
    )

    validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
        DIR,
        labels="inferred",
        label_mode="int",
        class_names=None,
        color_mode="rgb",
        batch_size=TEST_BATCH_SIZE,
        image_size=(96, 96),
        shuffle=True,
        seed=SEED,
        validation_split=0.2,
        subset="validation",
        interpolation="bilinear",
        follow_links=False,
    )

    return unlabeled_train_dataset, labeled_train_dataset, validation_dataset


unlabeled_train_dataset, labeled_train_dataset, validation_dataset = prepare_dataset()
def get_augmenter(
    height_factor,
    width_factor,
    color_jitter_rate,
    brightness_factor,
    contrast_factor,
    saturation_factor,
    hue_factor,
    rotation_rate,
    zoom,
    gb1,
    gb2,
    t_h,
    t_w,
):
    return keras.Sequential(
        [
            keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, IMAGE_CHANNELS)),
            keras_cv.layers.Rescaling(scale=1.0 / 255),
            keras_cv.layers.RandomFlip("horizontal"),
            keras_cv.layers.RandomCutout(
                height_factor, width_factor, fill_mode="constant",
            ),
            keras_cv.layers.RandomRotation(rotation_rate,fill_mode="reflect"
            ),
            keras_cv.layers.RandomApply(
                keras_cv.layers.RandomColorJitter(
                    value_range=(0, 1),
                    brightness_factor=brightness_factor,
                    contrast_factor=contrast_factor,
                    saturation_factor=saturation_factor,
                    hue_factor=hue_factor,
                ),
                rate=color_jitter_rate,
            ),
            keras_cv.layers.RandomApply(
                keras_cv.layers.RandomGaussianBlur(gb1,gb2),
                rate = 0.3
            ),
            tf.keras.layers.RandomTranslation(
                height_factor,
                width_factor,
                fill_mode="constant",
            )
        ]
    )
# Original Images
unlabeled_images = next(iter(unlabeled_train_dataset))
keras_cv.visualization.plot_image_gallery(
    images=unlabeled_images,
    value_range=(0, 255),
    rows=3,
    cols=3,
)

# Contrastive Augmentations
contrastive_augmenter = get_augmenter(**CONTRASTIVE_AUGMENTATION)
augmented_images = contrastive_augmenter(unlabeled_images)
keras_cv.visualization.plot_image_gallery(
    images=augmented_images,
    value_range=(0, 1),
    rows=3,
    cols=3,
)

# Classification Augmentations
classification_augmenter = get_augmenter(**CLASSIFICATION_AUGMENTATION)
augmented_images = classification_augmenter(unlabeled_images)
keras_cv.visualization.plot_image_gallery(
    images=augmented_images,
    value_range=(0, 1),
    rows=3,
    cols=3,
)

  
```

```python

with strategy.scope():     
     
    def get_encoder():  
        return keras.Sequential(  
        [  
            keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, IMAGE_CHANNELS)),  
            keras_cv.models.ResNet18Backbone(include_rescaling=False),
            keras.layers.GlobalAveragePooling2D(name="avg_pool"),  
        ],  
        name="encoder",  
    )

        
    class SimCLRTrainer(keras_cv.training.ContrastiveTrainer):    
        def __init__(self, encoder, augmenter, projector, probe=None, **kwargs):    
            super().__init__(    
                encoder=encoder,    
                augmenter=augmenter,    
                projector=projector,    
                probe=probe,    
                **kwargs,    
            )    
    
    simclr_model = SimCLRTrainer(    
    encoder=get_encoder(),    
    augmenter=get_augmenter(**CONTRASTIVE_AUGMENTATION),    
    projector=keras.Sequential(    
        [    
            keras.layers.Dense(16*PROJECTION_WIDTH, activation="relu", kernel_regularizer=regularizers.l2(0.01)),    
            keras.layers.Dense(16*PROJECTION_WIDTH, kernel_regularizer=regularizers.l2(0.01)),
            keras.layers.Dense(8*PROJECTION_WIDTH, kernel_regularizer=regularizers.l2(0.01)),
            keras.layers.BatchNormalization(),    
        ],    
        name="projector",    
    ),
    )   
    
    simclr_model.compile(    
        encoder_optimizer=keras.optimizers.Adam(),    
        encoder_loss=keras_cv.losses.SimCLRLoss(    
            temperature=TEMPERATURE,    
        ),    
    )    
    
    early_stopping = EarlyStopping(monitor='loss', patience=100)
    
    simclr_history = simclr_model.fit(
    unlabeled_train_dataset,
    epochs=500,
    callbacks=[early_stopping],
    )
    
    finetune_model = keras.Sequential(
    [
        keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, IMAGE_CHANNELS)),
        get_augmenter(**CLASSIFICATION_AUGMENTATION),
        simclr_model.encoder,
        keras.layers.Dense(1, activation="sigmoid"),
    ],
    name="finetuning_model",
    )
    
    finetune_model = keras.Sequential(
        [
            keras.Input(shape=(IMAGE_SIZE, IMAGE_SIZE, IMAGE_CHANNELS)),
            get_augmenter(**CLASSIFICATION_AUGMENTATION),
            simclr_model.encoder,
            keras.layers.Dense(1, activation="sigmoid", kernel_regularizer=regularizers.l2(0.01)),
        ],
        name="finetuning_model",
    )
    finetune_model.compile(
        optimizer=keras.optimizers.Adam(),
        loss=keras.losses.BinaryCrossentropy(from_logits=True),
        metrics=[keras.metrics.BinaryAccuracy(name="acc", dtype=tf.float32, threshold=0.5)],
    )
    early_stopping = EarlyStopping(monitor='val_loss', patience=50)
    
    finetune_history = finetune_model.fit(
        labeled_train_dataset, 
        epochs=300, 
        validation_data=validation_dataset,
        callbacks=[early_stopping]  # Add callbacks parameter
    )


```

```python
import os
import tensorflow as tf
import numpy as np
import keras
from keras.layers import Layer
from keras import initializers
from keras.utils import CustomObjectScope

class LayerScale(Layer):
    def __init__(self, init_values, projection_dim, **kwargs):
        super().__init__(**kwargs)
        self.init_values = init_values
        self.projection_dim = projection_dim

    def build(self, _):
        self.gamma = self.add_weight(
            shape=(self.projection_dim,),
            initializer=initializers.Constant(self.init_values),
            trainable=True,
        )

    def call(self, x):
        return x * self.gamma

    def get_config(self):
        config = super().get_config()
        config.update(
            {
                "init_values": self.init_values,
                "projection_dim": self.projection_dim,
            }
        )
        return config


class model:
    def __init__(self, path):
        self.model = tf.keras.models.load_model(os.path.join(path, 'SubmissionModel'), custom_objects={'LayerScale': LayerScale}, safe_mode=False)

    def predict(self, X):
        yhat = self.tta_prediction(X)
        return yhat

    def tta_prediction(self, image):
        images = list([image, tf.image.flip_left_right(image), tf.image.flip_up_down(image), tf.image.rot90(image, k=1), tf.image.rot90(image, k=3), tf.image.adjust_brightness(image, delta=0.2), tf.image.adjust_brightness(image, delta=-0.2),])

        preds = [self.model.predict(x, verbose=0) for x in images]
        mean_pred = tf.reduce_mean(preds, axis=0)

        tta_pred = tf.cast(mean_pred > 0.5, tf.int32)
        tta_pred_reshaped = tf.reshape(tta_pred, [-1])

        return tta_pred_reshaped
```
